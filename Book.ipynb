```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
plt.scatter(iris['petal_length'],iris['petal_width'])
plt.title('Scatter Plot')
plt.xlabel('No. of Users')
plt.ylabel('CPU Utilization')
plt.show()
```

```python
iris = iris.replace('setosa','Mexico')
iris = iris.replace('versicolor','Canada')
iris = iris.replace('virginica','US')
iris.head()
```

```python
sns.barplot(iris['species'],iris['sepal_length'])
plt.title('Bar Plot')
plt.xlabel('Country')
plt.ylabel('CPU Utilization')
plt.show()
```

```python
sns.lineplot(iris['sepal_length'],iris['sepal_width'])
plt.title('Line Plot')
plt.xlabel('No. of Users')
plt.ylabel('CPU Utilization')
plt.show()
```

```python
plt.hist(iris['sepal_width'], bins = 20)
plt.title('')
plt.title('Histogram')
plt.xlabel('CPU Utilization')
plt.ylabel('Frequency')
plt.show()
```

```python
labels = ['Admin', 'Read Access', 'Read/Write Access']
sizes = [10,60,30]
plt.pie(sizes, labels=labels, explode=(0.1,0.1,0.1))
plt.axis('equal')
plt.title('Pie Chart')
plt.show()
```

```python
sns.countplot(x='species', data=iris)
plt.title('Count Plot')
plt.xlabel('Country')
plt.show()
```

```python
sns.boxplot(x='sepal_length', data=iris)
plt.title('Box Plot')
plt.xlabel('No. of Users')
plt.show()
```

```python
sns.boxplot(x='species',y='sepal_length',data=iris)
plt.title('Box Plot Comparison with Multiple Series')
plt.xlabel('Country')
plt.ylabel('No. of Users')
plt.show()
```

```python
ax=sns.distplot(iris['sepal_width'],bins=25,kde=False)
ax.set(xlabel='CPU Utilization')
plt.show()
```

```python
ax=sns.distplot(iris['sepal_width'],bins=25,kde=True)
ax.set(xlabel='CPU Utilization')
plt.show()
```

```python
ax=sns.heatmap(iris.corr(),annot=True)
plt.show()
```

```python
ax=sns.jointplot(x='sepal_length',y='petal_length',data=iris,kind='scatter')
plt.show()
```

```python
grids = sns.PairGrid(iris)
ax=grids.map(plt.scatter)
ax.savefig('test.png')
```
